```python
import cv2
import numpy as np

# Load the color image
image = cv2.imread('bird.jpg')

# Convert image to float32 for precise calculations
image = np.float32(image)

# Step 1: Get Image Dimensions
height, width, channels = image.shape
total_pixels = height * width
print(f"Image Dimensions: Height={height}, Width={width}, Channels={channels}, Total Pixels={total_pixels}")

# Step 2: Compute Sum for Each Channel
sum_b = np.sum(image[:, :, 0])  # Sum of Blue channel
sum_g = np.sum(image[:, :, 1])  # Sum of Green channel
sum_r = np.sum(image[:, :, 2])  # Sum of Red channel

# Step 3: Compute Mean for Each Channel
mean_b = sum_b / total_pixels
mean_g = sum_g / total_pixels
mean_r = sum_r / total_pixels
print(f"Step 3 - Mean Calculation:")
print(f"Mean (B): {mean_b:.2f}, Mean (G): {mean_g:.2f}, Mean (R): {mean_r:.2f}")

# Step 4: Compute Variance for Each Channel
sum_squared_diff_b = np.sum((image[:, :, 0] - mean_b) ** 2)
sum_squared_diff_g = np.sum((image[:, :, 1] - mean_g) ** 2)
sum_squared_diff_r = np.sum((image[:, :, 2] - mean_r) ** 2)

var_b = sum_squared_diff_b / total_pixels
var_g = sum_squared_diff_g / total_pixels
var_r = sum_squared_diff_r / total_pixels
print(f"Step 4 - Variance Calculation:")
print(f"Variance (B): {var_b:.2f}, Variance (G): {var_g:.2f}, Variance (R): {var_r:.2f}")

# Step 5: Compute Overall Mean and Variance
overall_mean = np.mean(image)
overall_variance = np.var(image)

print(f"Step 5 - Overall Mean and Variance:")
print(f"Overall Mean: {overall_mean:.2f}")
print(f"Overall Variance: {overall_variance:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a 3x3 color image (B, G, R values)
image = np.array([
    [[100, 120, 130], [110, 140, 150], [120, 160, 170]],
    [[130, 150, 160], [140, 170, 180], [150, 180, 190]],
    [[160, 180, 190], [170, 190, 200], [180, 200, 210]]
], dtype=np.float32)

# Compute mean and variance for each channel
mean_b, mean_g, mean_r = np.mean(image, axis=(0, 1))
var_b, var_g, var_r = np.var(image, axis=(0, 1))

# Create labels for channels
labels = ['Blue', 'Green', 'Red']

# Create a figure with subplots
plt.figure(figsize=(12, 5))

# Plot Mean Values (Bar Chart)
plt.subplot(1, 2, 1)
plt.bar(labels, [mean_b, mean_g, mean_r], color=['blue', 'green', 'red'])
plt.title("Mean Pixel Intensity")
plt.ylabel("Mean Value")
plt.ylim(0, 255)  # Pixel intensity range

# Plot Variance Values (Bar Chart)
plt.subplot(1, 2, 2)
plt.bar(labels, [var_b, var_g, var_r], color=['blue', 'green', 'red'])
plt.title("Variance of Pixel Intensities")
plt.ylabel("Variance Value")

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the real image (bird.jpg)
image = cv2.imread('bird.jpg')

# Convert image from BGR to RGB (Matplotlib expects RGB format)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Extract individual color channels
blue_channel = image[:, :, 0].flatten()
green_channel = image[:, :, 1].flatten()
red_channel = image[:, :, 2].flatten()

# Create a figure with subplots
plt.figure(figsize=(12, 4))

# Histogram for Blue Channel
plt.subplot(1, 3, 1)
plt.hist(blue_channel, bins=50, color='blue', alpha=0.7, edgecolor='black')
plt.title("Blue Channel Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.xlim(0, 255)

# Histogram for Green Channel
plt.subplot(1, 3, 2)
plt.hist(green_channel, bins=50, color='green', alpha=0.7, edgecolor='black')
plt.title("Green Channel Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.xlim(0, 255)

# Histogram for Red Channel
plt.subplot(1, 3, 3)
plt.hist(red_channel, bins=50, color='red', alpha=0.7, edgecolor='black')
plt.title("Red Channel Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.xlim(0, 255)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the real image (bird.jpg)
image = cv2.imread('bird.jpg')

# Convert image from BGR to RGB (Matplotlib expects RGB format)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Extract individual color channels
blue_channel = image[:, :, 0].flatten()
green_channel = image[:, :, 1].flatten()
red_channel = image[:, :, 2].flatten()

# Compute mean and variance for each channel
mean_b, var_b = np.mean(blue_channel), np.var(blue_channel)
mean_g, var_g = np.mean(green_channel), np.var(green_channel)
mean_r, var_r = np.mean(red_channel), np.var(red_channel)

# Print results
print(f"Mean (B, G, R): {mean_b:.2f}, {mean_g:.2f}, {mean_r:.2f}")
print(f"Variance (B, G, R): {var_b:.2f}, {var_g:.2f}, {var_r:.2f}")

# Create a figure with subplots
plt.figure(figsize=(12, 5))

# Bar Chart for Mean
plt.subplot(1, 2, 1)
plt.bar(['Blue', 'Green', 'Red'], [mean_b, mean_g, mean_r], color=['blue', 'green', 'red'])
plt.title("Mean Pixel Intensity")
plt.ylabel("Mean Value")
plt.ylim(0, 255)  # Pixel intensity range

# Bar Chart for Variance
plt.subplot(1, 2, 2)
plt.bar(['Blue', 'Green', 'Red'], [var_b, var_g, var_r], color=['blue', 'green', 'red'])
plt.title("Variance of Pixel Intensities")
plt.ylabel("Variance Value")

# Show the plots
plt.tight_layout()
plt.show()
```
